### définition explicite de G:

### définition implicite de G
une definition de gamma implicite, par les successeurs d'un ou plusieurs elements "initiaux" parmis les sommets  
et les successeurs des successeurs, et ainsi de suite.
$$\Gamma : S \rightarrow (S * N^*)^k$$
$$ e_i \rightarrow (\{e_j * c_{ij}\})$$
et
$$G : ( \{e_i\} \subset S,\cup \tau (elem, (\tau (\Gamma, \{e_i\} ))), \cup  \tau (elem, (\tau (\Gamma, \{\tau (elem, (\tau (\Gamma, \{e_i\} )))\} ))), ... )$$ 
c est à dire les elements de depart, et leurs successeurs ,et les successeurs de leurs successeurs... récursivement

avec l applicateur d une methode unaire sur un ensemble ("map") $$ \tau : A^n x f \rightarrow (f(a_1),....,f(a_n)) $$  
et l accesseurs aux elements à partir du doublet :$$ elem : (S * N^*) \rightarrow N^* $$
$$ (e_i * N^*) \rightarrow e_i $$
